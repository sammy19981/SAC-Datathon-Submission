```python
"""
LEHD LODE Transit Gap Analysis
================================
Goal: Identify where new public transit lines should go by finding:
  1. Job center hotspots (where people work)
  2. Worker origin clusters (where people live)
  3. High-demand corridors with no good transit connection ("transit gaps")

LODE Columns Used:
  w_geocode  - Census block where job is located
  h_geocode  - Census block where worker lives
  S000       - Total jobs (all workers)
  SA01       - Jobs held by workers age 29 or younger
  SA02       - Jobs held by workers age 30–54
  SA03       - Jobs held by workers age 55+
  SE01       - Jobs with monthly earnings <= $1,250 (low income)
  SE02       - Jobs with monthly earnings $1,251–$3,333 (middle income)
  SE03       - Jobs with monthly earnings > $3,333 (high income)
  SI01       - Jobs in Goods Producing industries
  SI02       - Jobs in Trade, Transportation, Utilities
  SI03       - Jobs in All Other Services
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import time
import os

def timer(label, t):
    print(f"  ✓ {label} ({time.time()-t:.1f}s)")
    return time.time()

# ─────────────────────────────────────────
# STEP 1: LOAD & VALIDATE DATA
# ─────────────────────────────────────────
t = time.time()
file_path = "census_data.csv"
df = pd.read_csv(file_path, dtype={'w_geocode': str, 'h_geocode': str})

# Force numeric job columns in one shot
job_cols = ['S000', 'SA01', 'SA02', 'SA03', 'SE01', 'SE02', 'SE03', 'SI01', 'SI02', 'SI03']
df[job_cols] = df[job_cols].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=['w_geocode', 'h_geocode', 'S000'])
print(f"Loaded {len(df):,} origin-destination job flows")
t = timer("Load & validate", t)


# ─────────────────────────────────────────
# STEP 2: LOAD REAL COORDINATES FROM TIGER SHAPEFILE
#
# tl_2025_24_tabblock20.shp provides actual Census block centroids.
# GEOID column is GEOID20 — a 15-char string matching LODES geocodes.
# ─────────────────────────────────────────
import geopandas as gpd

shapefile_path = "tl_2025_24_tabblock20.shp"
print("Loading shapefile (may take ~30s for a full state)...")
blocks = gpd.read_file(shapefile_path)

# Project to a meter-based CRS for accurate centroids, then back to WGS84
blocks_proj = blocks.to_crs(epsg=32618)  # UTM zone 18N covers Maryland
blocks['lat'] = blocks_proj.geometry.centroid.to_crs(epsg=4326).y
blocks['lon'] = blocks_proj.geometry.centroid.to_crs(epsg=4326).x

block_coords = blocks[['GEOID20', 'lat', 'lon']].copy()

# Normalize both sides: strip whitespace, find right pad length, then zfill
shp_sample   = str(block_coords['GEOID20'].iloc[0]).strip()
lodes_sample = str(df['w_geocode'].iloc[0]).strip()
pad_len = max(len(shp_sample), len(lodes_sample), 15)
print(f"  Raw shapefile GEOID sample : {shp_sample!r} (len={len(shp_sample)})")
print(f"  Raw LODES w_geocode sample : {lodes_sample!r} (len={len(lodes_sample)})")
print(f"  Padding both to {pad_len} chars")

block_coords['GEOID20'] = block_coords['GEOID20'].astype(str).str.strip().str.zfill(pad_len)
df['w_geocode']         = df['w_geocode'].astype(str).str.strip().str.zfill(pad_len)
df['h_geocode']         = df['h_geocode'].astype(str).str.strip().str.zfill(pad_len)

overlap_w = set(block_coords['GEOID20']).intersection(set(df['w_geocode']))
print(f"  Matching w_geocodes after normalization: {len(overlap_w):,} of {df['w_geocode'].nunique():,} unique work blocks")
t = timer("Shapefile load", t)

# Merge real coords onto LODES rows (work and home blocks)
df = df.merge(
    block_coords.rename(columns={'GEOID20': 'w_geocode', 'lat': 'wlat', 'lon': 'wlon'}),
    on='w_geocode', how='left'
)
df = df.merge(
    block_coords.rename(columns={'GEOID20': 'h_geocode', 'lat': 'hlat', 'lon': 'hlon'}),
    on='h_geocode', how='left'
)

# Only require the WORK block to have coords (it must be in MD).
# Home blocks may be out-of-state — keep those rows but null their coords.
missing_work = df['wlat'].isna().sum()
missing_home = df['hlat'].isna().sum()
if missing_work > 0:
    print(f"  ⚠ {missing_work:,} rows missing work block coords — dropping")
    df = df.dropna(subset=['wlat', 'wlon'])
if missing_home > 0:
    print(f"  ℹ {missing_home:,} rows have out-of-state home blocks — kept, but excluded from corridor distance")
    # Fill missing home coords with work coords as fallback (0-mile commute placeholder)
    df['hlat'] = df['hlat'].fillna(df['wlat'])
    df['hlon'] = df['hlon'].fillna(df['wlon'])
df['wx'] = df['wlon']
df['wy'] = df['wlat']
df['hx'] = df['hlon']
df['hy'] = df['hlat']

t = timer("Coordinate merge", t)


# ─────────────────────────────────────────
# HAVERSINE DISTANCE (miles) — vectorized
# Replaces the fake GEOID-space distance used previously.
# ─────────────────────────────────────────
def haversine_miles(lat1, lon1, lat2, lon2):
    R = 3958.8
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    return R * 2 * np.arcsin(np.sqrt(a))


# ─────────────────────────────────────────
# STEP 3: BUILD JOB CENTER PROFILES
#
# Aggregate all flows arriving at each work block.
# This tells us: how many workers come here, from how far,
# and what income level they are.
# ─────────────────────────────────────────
job_centers = df.groupby('w_geocode').agg(
    total_jobs    = ('S000',  'sum'),
    low_income    = ('SE01',  'sum'),   # workers most dependent on transit
    mid_income    = ('SE02',  'sum'),
    high_income   = ('SE03',  'sum'),
    young_workers = ('SA01',  'sum'),
    wx            = ('wx',    'first'),
    wy            = ('wy',    'first'),
    unique_origins= ('h_geocode', 'nunique'),  # how many neighborhoods feed this job center
).reset_index()

# Transit need score: weight low-income workers more heavily
# (they are most transit-dependent) and penalize for spread of origins
# (many origins = more people need transit)
job_centers['transit_need'] = (
    job_centers['low_income'] * 2.0 +
    job_centers['mid_income'] * 1.0 +
    job_centers['unique_origins'] * 0.5
)

print(f"\nTop 10 job centers by transit need:")
print(job_centers.nlargest(10, 'transit_need')[
    ['w_geocode','total_jobs','low_income','unique_origins','transit_need']
])
t = timer("Job center profiles", t)


# ─────────────────────────────────────────
# STEP 4: CLUSTER JOB CENTERS
#
# Group spatially close job centers into transit-relevant zones.
# We cluster on: location + transit need score.
# This finds "job hubs" that a single transit line could serve.
# ─────────────────────────────────────────
# Filter to meaningful job centers only
significant = job_centers[job_centers['total_jobs'] >= job_centers['total_jobs'].quantile(0.5)].copy()

cluster_features = significant[['wx', 'wy', 'transit_need']].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cluster_features)

k = 6  # number of job hub zones (tune based on your metro area)
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
significant['job_hub'] = kmeans.fit_predict(X_scaled)

# Merge cluster labels back
job_centers = job_centers.merge(
    significant[['w_geocode', 'job_hub']], on='w_geocode', how='left'
)
job_centers['job_hub'] = job_centers['job_hub'].fillna(-1).astype(int)
t = timer("Clustering", t)


# ─────────────────────────────────────────
# STEP 5: FIND HIGH-DEMAND CORRIDORS
#
# Aggregate block-level flows to Census TRACT pairs.
# Tract GEOID = first 11 chars of block GEOID (state 2 + county 3 + tract 6).
# Tracts have ~4,000 residents — large enough to justify transit investment.
# ─────────────────────────────────────────

# Extract tract IDs from block GEOIDs
df['h_tract'] = df['h_geocode'].str[:11]
df['w_tract'] = df['w_geocode'].str[:11]

# Pre-compute weighted coord columns for vectorized aggregation
df['hx_w'] = df['hx'] * df['S000']
df['hy_w'] = df['hy'] * df['S000']
df['wx_w'] = df['wx'] * df['S000']
df['wy_w'] = df['wy'] * df['S000']

# Vectorized groupby — ~50x faster than apply(lambda)
grp = df.groupby(['h_tract', 'w_tract'])
corridors = grp.agg(
    total_flow  = ('S000',  'sum'),
    low_income  = ('SE01',  'sum'),
    hx_wsum     = ('hx_w', 'sum'),
    hy_wsum     = ('hy_w', 'sum'),
    wx_wsum     = ('wx_w', 'sum'),
    wy_wsum     = ('wy_w', 'sum'),
    h_geocode   = ('h_geocode', 'first'),
    w_geocode   = ('w_geocode', 'first'),
).reset_index()

# Compute weighted centroids
corridors['hx'] = corridors['hx_wsum'] / corridors['total_flow']
corridors['hy'] = corridors['hy_wsum'] / corridors['total_flow']
corridors['wx'] = corridors['wx_wsum'] / corridors['total_flow']
corridors['wy'] = corridors['wy_wsum'] / corridors['total_flow']
corridors = corridors.drop(columns=['hx_wsum','hy_wsum','wx_wsum','wy_wsum'])

# Filter to meaningful flows — at least 50 workers per tract corridor
# This removes noise and ensures corridors represent real demand
corridors = corridors[corridors['total_flow'] >= 50].copy()

# Weighted demand: low-income workers count more
corridors['weighted_demand'] = corridors['low_income'] * 2 + corridors['total_flow']

# Real straight-line distance between tract centroids
corridors['distance_miles'] = haversine_miles(
    corridors['hy'].values, corridors['hx'].values,
    corridors['wy'].values, corridors['wx'].values
)

# Drop same-tract and out-of-state placeholder rows
corridors = corridors[
    (corridors['h_tract'] != corridors['w_tract']) &
    (corridors['distance_miles'] > 0.5)
].copy()

# Transit gap score
corridors['gap_score'] = corridors['weighted_demand'] * np.log1p(corridors['distance_miles'])

print(f"  {len(corridors):,} tract-level corridors")
print(f"  Workers per corridor — median: {corridors['total_flow'].median():.0f}, "
      f"mean: {corridors['total_flow'].mean():.0f}, "
      f"max: {corridors['total_flow'].max():.0f}")

top_corridors = corridors.nlargest(500, 'gap_score')

print(f"\nTop 10 transit gap corridors (tract level):")
print(top_corridors[['h_tract','w_tract','total_flow','low_income','distance_miles','gap_score']].head(10))
t = timer("Corridor analysis", t)


# ─────────────────────────────────────────
# STEP 6: VISUALIZATION — SINGLE MAP WITH GTFS OVERLAY
# ─────────────────────────────────────────
import contextily as ctx
import geopandas as gpd
from shapely.geometry import Point, LineString
import zipfile

CRS_DISPLAY = "EPSG:3857"
CRS_DATA    = "EPSG:4326"

def to_webmerc(df, lon_col, lat_col):
    gdf = gpd.GeoDataFrame(
        df,
        geometry=[Point(xy) for xy in zip(df[lon_col], df[lat_col])],
        crs=CRS_DATA
    )
    return gdf.to_crs(CRS_DISPLAY)

# ─────────────────────────────────────────
# LOAD GTFS TRANSIT FEEDS
# Each feed gets its own color on the map
# ─────────────────────────────────────────
gtfs_feeds = [
    ("mdotmta_gtfs_commuterbus", "Commuter Bus",  "#E57C00"),  # dark orange
    ("mdotmta_gtfs_lightrail",   "Light Rail",    "#9B30FF"),  # purple
    ("mdotmta_gtfs_marc",        "MARC Train",    "#006400"),  # dark green
    ("mdotmta_gtfs_metro",       "Metro Subway",  "#CC0000"),  # dark red
]

def load_gtfs_shapes(zip_path, crs_out):
    with zipfile.ZipFile(zip_path) as z:
        with z.open("shapes.txt") as f:
            shapes = pd.read_csv(f)
    shapes = shapes.sort_values(['shape_id', 'shape_pt_sequence'])
    # Build LineStrings explicitly to avoid Shapely deprecation warning
    geoms = []
    ids   = []
    for shape_id, grp in shapes.groupby('shape_id'):
        coords = list(zip(grp['shape_pt_lon'], grp['shape_pt_lat']))
        if len(coords) >= 2:
            geoms.append(LineString(coords))
            ids.append(shape_id)
    gdf = gpd.GeoDataFrame({'shape_id': ids, 'geometry': geoms}, crs=CRS_DATA)
    return gdf.to_crs(crs_out)

loaded_feeds = []
for filename, label, color in gtfs_feeds:
    try:
        gdf = load_gtfs_shapes(f"{filename}.zip", CRS_DISPLAY)
        # Deduplicate shapes so overlapping routes don't overdraw
        gdf = gdf.drop_duplicates(subset='geometry')
        loaded_feeds.append((gdf, label, color))
        print(f"  ✓ {label}: {len(gdf):,} route shapes")
    except Exception as e:
        print(f"  ⚠ Could not load {label} ({filename}.zip): {e}")

# ── Load local bus routes from bus_routes/ folder ──────────────────
# These are combined with GTFS for the proximity filter so corridors
# already served by local bus are correctly excluded from "unserved".
BUS_FOLDER = "bus_routes"
SKIP_RAIL  = ["marc", "light rail", "lightrail", "metro subway", "amtrak", "subway"]

def label_bus_file(fname):
    fl = fname.lower()
    if "wmata" in fl or "metrobus" in fl:
        return "WMATA Metrobus", "#CC0000"
    if "mta" in fl:
        return "MTA Commuter Bus", "#E57C00"
    return "Local Bus", "#4A90D9"

if os.path.isdir(BUS_FOLDER):
    print(f"\nLoading local bus routes from {BUS_FOLDER}/...")
    for fname in sorted(os.listdir(BUS_FOLDER)):
        if not (fname.lower().endswith(".geojson") or fname.lower().endswith(".json")):
            continue
        if any(kw in fname.lower() for kw in SKIP_RAIL):
            continue
        fpath = os.path.join(BUS_FOLDER, fname)
        try:
            gdf = gpd.read_file(fpath)
            if gdf.crs is None:
                gdf = gdf.set_crs(CRS_DATA)
            else:
                gdf = gdf.to_crs(CRS_DATA)
            lines = gdf[gdf.geometry.geom_type.isin(
                ["LineString", "MultiLineString"])].copy()
            if len(lines) == 0:
                lines = gdf.copy()
            if len(lines) == 0:
                continue
            gdf_disp = lines.to_crs(CRS_DISPLAY)
            label, color = label_bus_file(fname)
            loaded_feeds.append((gdf_disp, label, color))
            print(f"  ✓ {fname}: {len(gdf_disp):,} shapes ({label})")
        except Exception as e:
            print(f"  ⚠ {fname}: {e}")
else:
    print(f"\n  ℹ No {BUS_FOLDER}/ folder found — skipping local bus routes")

# ─────────────────────────────────────────
# SINGLE MAP: Gap corridors + GTFS routes
# ─────────────────────────────────────────
fig, ax = plt.subplots(1, 1, figsize=(14, 16), subplot_kw={'aspect': 'equal'})
ax.set_title("Maryland Top 500 Underserved Commute Corridors vs Existing Transit",
             fontsize=13, fontweight='bold', pad=12)

cmap    = plt.cm.YlOrRd
max_gap = top_corridors['gap_score'].max() if not top_corridors.empty else 1

# Draw each GTFS feed behind the corridors with its own color
for gdf, label, color in loaded_feeds:
    gdf.plot(ax=ax, color=color, linewidth=1.2, alpha=0.7, zorder=2, label=label)

# Draw gap corridors
tc       = top_corridors.copy()
home_gdf = to_webmerc(tc, 'hx', 'hy')
work_gdf = to_webmerc(tc, 'wx', 'wy')

for i, (_, row) in enumerate(tc.iterrows()):
    norm_score = row['gap_score'] / max_gap
    color = cmap(norm_score)
    lw    = 0.5 + norm_score * 3
    ax.plot(
        [home_gdf.geometry.iloc[i].x, work_gdf.geometry.iloc[i].x],
        [home_gdf.geometry.iloc[i].y, work_gdf.geometry.iloc[i].y],
        color=color, linewidth=lw, alpha=0.8, zorder=3
    )

# Home and work markers
ax.scatter(home_gdf.geometry.x, home_gdf.geometry.y,
           c='steelblue', s=15, alpha=0.7, zorder=4, label='Home area')
ax.scatter(work_gdf.geometry.x, work_gdf.geometry.y,
           c='crimson',   s=15, alpha=0.7, zorder=4, label='Job center')

# Basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, zoom='auto')

# Colorbar for gap score
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(0, max_gap))
sm.set_array([])
plt.colorbar(sm, ax=ax, label='Transit Gap Score', shrink=0.5, pad=0.01)

ax.legend(fontsize=9, loc='upper right')
ax.set_axis_off()

plt.tight_layout()
plt.savefig("transit_gap_analysis.png", dpi=150, bbox_inches='tight')
plt.show()
print("\nPlot saved to transit_gap_analysis.png")


# ─────────────────────────────────────────
# STEP 7: TRANSIT PROXIMITY FILTER
#
# For each corridor midpoint, check whether any existing
# transit route passes within 1 mile. Corridors with no
# nearby transit are genuinely unserved — highest priority.
# Runs after map draw so to_webmerc and home_gdf/work_gdf are available.
# ─────────────────────────────────────────
PROXIMITY_MILES   = 1.0
PROXIMITY_METERS  = PROXIMITY_MILES * 1609.34

print("\nChecking transit proximity for top corridors...")

if loaded_feeds:
    from shapely.ops import unary_union as shapely_unary_union

    # Merge all transit geometries into one for fast distance lookup
    all_transit_geoms = pd.concat(
        [gdf[['geometry']] for gdf, _, _ in loaded_feeds], ignore_index=True
    )
    transit_union = shapely_unary_union(all_transit_geoms['geometry'].values)

    # Reuse home_gdf / work_gdf already computed during map drawing
    midpoints = gpd.GeoDataFrame(
        geometry=[
            Point((h.x + w.x) / 2, (h.y + w.y) / 2)
            for h, w in zip(home_gdf.geometry, work_gdf.geometry)
        ],
        crs=CRS_DISPLAY
    )

    tc_check = top_corridors.copy().reset_index(drop=True)
    tc_check['dist_to_transit_m']     = midpoints.geometry.apply(transit_union.distance)
    tc_check['dist_to_transit_miles'] = tc_check['dist_to_transit_m'] / 1609.34
    tc_check['has_nearby_transit']    = tc_check['dist_to_transit_m'] <= PROXIMITY_METERS

    served   = tc_check['has_nearby_transit'].sum()
    unserved = (~tc_check['has_nearby_transit']).sum()
    print(f"  {served} corridors within {PROXIMITY_MILES} mile of existing transit")
    print(f"  {unserved} corridors with NO nearby transit — genuinely unserved")

    unserved_corridors = tc_check[~tc_check['has_nearby_transit']].sort_values(
        'gap_score', ascending=False
    )

    # Highlight unserved corridors on the map with a white outline
    uc_idx  = unserved_corridors.index.tolist()
    for i in uc_idx:
        ax.plot(
            [home_gdf.geometry.iloc[i].x, work_gdf.geometry.iloc[i].x],
            [home_gdf.geometry.iloc[i].y, work_gdf.geometry.iloc[i].y],
            color='white', linewidth=2.5, alpha=0.5, zorder=2, solid_capstyle='round'
        )

    plt.savefig("transit_gap_analysis.png", dpi=150, bbox_inches='tight')
    print("  Map updated with unserved corridor highlights (white outline)")

    # ── SECOND MAP: unserved corridors only ──────────────────────────
    fig2, ax2b = plt.subplots(1, 1, figsize=(14, 16), subplot_kw={'aspect': 'equal'})
    ax2b.set_title("Maryland Genuinely Unserved Commute Corridors\n(>1 mile from any existing transit)",
                   fontsize=13, fontweight='bold', pad=12)

    # Draw transit routes for context
    for gdf, label, color in loaded_feeds:
        gdf.plot(ax=ax2b, color=color, linewidth=1.2, alpha=0.7, zorder=2, label=label)

    # Draw ONLY unserved corridors
    uc_home2 = to_webmerc(unserved_corridors.reset_index(drop=True), 'hx', 'hy')
    uc_work2 = to_webmerc(unserved_corridors.reset_index(drop=True), 'wx', 'wy')
    uc_max_gap = unserved_corridors['gap_score'].max()

    for i in range(len(unserved_corridors)):
        norm_score = unserved_corridors['gap_score'].iloc[i] / uc_max_gap
        color = cmap(norm_score)
        lw    = 0.5 + norm_score * 3
        ax2b.plot(
            [uc_home2.geometry.iloc[i].x, uc_work2.geometry.iloc[i].x],
            [uc_home2.geometry.iloc[i].y, uc_work2.geometry.iloc[i].y],
            color=color, linewidth=lw, alpha=0.9, zorder=3
        )

    ax2b.scatter(uc_home2.geometry.x, uc_home2.geometry.y,
                 c='steelblue', s=15, alpha=0.8, zorder=4, label='Home area')
    ax2b.scatter(uc_work2.geometry.x, uc_work2.geometry.y,
                 c='crimson',   s=15, alpha=0.8, zorder=4, label='Job center')

    ctx.add_basemap(ax2b, source=ctx.providers.CartoDB.Positron, zoom='auto')

    sm2 = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(0, uc_max_gap))
    sm2.set_array([])
    plt.colorbar(sm2, ax=ax2b, label='Transit Gap Score', shrink=0.5, pad=0.01)

    ax2b.legend(fontsize=9, loc='upper right')
    ax2b.set_axis_off()

    plt.tight_layout()
    plt.savefig("transit_gaps_unserved_only.png", dpi=150, bbox_inches='tight')
    plt.show()
    print("  Unserved-only map saved to transit_gaps_unserved_only.png")

else:
    print("  No GTFS feeds loaded — skipping proximity filter")
    unserved_corridors = top_corridors.copy()
    tc_check = top_corridors.copy()
    tc_check['dist_to_transit_miles'] = float('nan')
    tc_check['has_nearby_transit']    = False


# ─────────────────────────────────────────
# STEP 8: PRIORITY REPORT
# ─────────────────────────────────────────
print("\n" + "="*60)
print("TRANSIT PRIORITY REPORT")
print("="*60)

hub_summary = significant.groupby('job_hub').agg(
    total_jobs   = ('total_jobs', 'sum'),
    low_income   = ('low_income', 'sum'),
    blocks       = ('w_geocode', 'count'),
    transit_need = ('transit_need', 'sum'),
).sort_values('transit_need', ascending=False)

print("\nJob Hub Rankings (highest transit need first):")
print(hub_summary.to_string())

print(f"\nTop 20 UNSERVED tract corridors (no transit within {PROXIMITY_MILES} mile):")
print(f"{'Rank':<5} {'Workers':>7} {'Low-Inc':>7} {'Miles':>6} {'Gap Score':>10} {'Dist to Transit':>16}  Home Tract → Work Tract")
print("-" * 95)
county_fips = {
    "001":"Allegany","003":"Anne Arundel","005":"Baltimore Co.",
    "009":"Calvert","011":"Caroline","013":"Carroll","015":"Cecil",
    "017":"Charles","019":"Dorchester","021":"Frederick","023":"Garrett",
    "025":"Harford","027":"Howard","029":"Kent","031":"Montgomery",
    "033":"Prince George's","035":"Queen Anne's","037":"St. Mary's",
    "039":"Somerset","041":"Talbot","043":"Washington",
    "045":"Wicomico","047":"Worcester","510":"Baltimore City"
}
def tract_label(tract):
    t = str(tract).zfill(11)
    county = county_fips.get(t[2:5], t[2:5])
    return f"{county} Tract {t[5:]}"

for rank, (_, row) in enumerate(unserved_corridors.head(20).iterrows(), 1):
    print(f"  {rank:<3} {int(row['total_flow']):>7} {int(row['low_income']):>7} "
          f"{row['distance_miles']:>6.1f} {row['gap_score']:>10.1f} "
          f"{row['dist_to_transit_miles']:>14.1f}mi  "
          f"{tract_label(row['h_tract'])} → {tract_label(row['w_tract'])}")

# Export unserved corridors to CSV for further planning use
unserved_corridors.to_csv("unserved_corridors.csv", index=False)
print("\nFull unserved corridor list saved to unserved_corridors.csv")


# ─────────────────────────────────────────
# STEP 9: STATEWIDE COMMUTE DISTANCE HEATMAPS
#
# Two maps:
#   LEFT  — home blocks: where do people with long commutes live?
#   RIGHT — work blocks: which job centers are hardest to reach?
#
# Method: aggregate average commute distance per Census block
# from the full 2M-row corridors DataFrame, project to Web Mercator,
# then interpolate to a raster grid using scipy and render with imshow.
# ─────────────────────────────────────────
from scipy.interpolate import griddata

print("\nBuilding statewide commute distance heatmaps...")

# ── Aggregate average commute distance per home block ──
# Use full df (all flows) not filtered corridors for complete coverage
all_flows = df[df['distance_miles'] > 0.1].copy() if 'distance_miles' in df.columns else df.copy()
if 'distance_miles' not in all_flows.columns:
    all_flows['distance_miles'] = haversine_miles(
        all_flows['hy'].values, all_flows['hx'].values,
        all_flows['wy'].values, all_flows['wx'].values
    )
    all_flows = all_flows[all_flows['distance_miles'] > 0.1]

home_heat = (
    all_flows
    .groupby(['h_geocode', 'hx', 'hy'])
    .apply(lambda g: np.average(g['distance_miles'], weights=g['S000']))
    .reset_index()
    .rename(columns={0: 'avg_commute'})
)

# ── Aggregate average commute distance per work block ──
work_heat = (
    all_flows
    .groupby(['w_geocode', 'wx', 'wy'])
    .apply(lambda g: np.average(g['distance_miles'], weights=g['S000']))
    .reset_index()
    .rename(columns={0: 'avg_commute'})
)

print(f"  Home blocks: {len(home_heat):,} | Work blocks: {len(work_heat):,}")

def build_heatmap_raster(lon, lat, values, md_boundary=None, resolution=400):
    from rasterio.transform import from_bounds
    from rasterio.features import geometry_mask
    pts = gpd.GeoDataFrame(
        geometry=gpd.points_from_xy(lon, lat), crs=CRS_DATA
    ).to_crs(CRS_DISPLAY)
    x = pts.geometry.x.values
    y = pts.geometry.y.values
    xmin, xmax = x.min(), x.max()
    ymin, ymax = y.min(), y.max()
    xi = np.linspace(xmin, xmax, resolution)
    yi = np.linspace(ymin, ymax, resolution)
    xi_grid, yi_grid = np.meshgrid(xi, yi)
    # Linear for smooth interior, nearest to fill convex-hull edges
    grid_lin = griddata((x, y), values, (xi_grid, yi_grid), method='linear')
    grid_near = griddata((x, y), values, (xi_grid, yi_grid), method='nearest')
    grid = np.where(np.isnan(grid_lin), grid_near, grid_lin)
    # Clip to MD boundary using rasterio (very fast rasterization)
    if md_boundary is not None:
        transform = from_bounds(xmin, ymin, xmax, ymax, resolution, resolution)
        md_geom = md_boundary.union_all()
        mask = geometry_mask([md_geom.__geo_interface__],
                             transform=transform,
                             invert=True,
                             out_shape=(resolution, resolution))
        # rasterio origin is top-left, flip to match imshow lower origin
        mask = np.flipud(mask)
        grid = np.where(mask, grid, np.nan)
    return xi, yi, grid, xmin, xmax, ymin, ymax

# ── Load Maryland state boundary for clipping ──
# Uses the Census state-level shapefile (single polygon, loads instantly)
# Download: https://www2.census.gov/geo/tiger/TIGER2025/STATE/tl_2025_us_state.zip
# Or we fetch just Maryland from the Census cartographic boundary API
import urllib.request, zipfile, io, os

md_state_path = "cb_2024_24_state_500k.shp"
if not os.path.exists(md_state_path):
    print("  Downloading Maryland shoreline boundary (~500KB)...")
    # Cartographic boundary = clipped to coastline, excludes water bodies
    url = "https://www2.census.gov/geo/tiger/GENZ2024/shp/cb_2024_us_state_500k.zip"
    try:
        with urllib.request.urlopen(url, timeout=30) as r:
            zdata = r.read()
        with zipfile.ZipFile(io.BytesIO(zdata)) as z:
            z.extractall("cb_states")
        us_states = gpd.read_file("cb_states/cb_2024_us_state_500k.shp")
        md_only = us_states[us_states['STATEFP'] == '24']
        md_only.to_file(md_state_path)
        print("  Maryland shoreline boundary downloaded and saved")
    except Exception as e:
        print(f"  ⚠ Download failed ({e}) — clipping disabled")
        md_state = None

if os.path.exists(md_state_path):
    md_state = gpd.read_file(md_state_path).to_crs(CRS_DISPLAY)
    print("  Maryland shoreline boundary loaded")

# ── Build rasters ──
print("  Interpolating home-side raster (this may take ~30s)...")
hxi, hyi, h_grid, hxmin, hxmax, hymin, hymax = build_heatmap_raster(
    home_heat['hx'].values, home_heat['hy'].values,
    home_heat['avg_commute'].values, md_state
)

print("  Interpolating work-side raster...")
wxi, wyi, w_grid, wxmin, wxmax, wymin, wymax = build_heatmap_raster(
    work_heat['wx'].values, work_heat['wy'].values,
    work_heat['avg_commute'].values, md_state
)

# ── Plot ──
fig3, (ax_h, ax_w) = plt.subplots(1, 2, figsize=(22, 12),
                                   subplot_kw={'aspect': 'equal'})
fig3.suptitle("Maryland Statewide Commute Distance Heatmap",
              fontsize=15, fontweight='bold')
fig3.subplots_adjust(top=0.93)

heat_cmap = plt.cm.plasma   # yellow=short, purple=long

def plot_heatmap(ax, xi, yi, grid, xmin, xmax, ymin, ymax,
                 title, cbar_label, md_boundary=None):
    # Use full state boundary extent so nothing is clipped
    if md_boundary is not None:
        bx = md_boundary.total_bounds  # [xmin, ymin, xmax, ymax]
        pad = 20000  # 20km padding in meters
        ax.set_xlim(bx[0] - pad, bx[2] + pad)
        ax.set_ylim(bx[1] - pad, bx[3] + pad)
    else:
        ax.set_xlim(xmin, xmax)
        ax.set_ylim(ymin, ymax)
    ctx.add_basemap(ax, source=ctx.providers.CartoDB.PositronNoLabels, zoom='auto')

    im = ax.imshow(
        grid,
        origin='lower',
        extent=[xmin, xmax, ymin, ymax],
        cmap=heat_cmap,
        alpha=0.7,
        aspect='auto',
        vmin=np.nanpercentile(grid[~np.isnan(grid)], 5),
        vmax=np.nanpercentile(grid[~np.isnan(grid)], 95),
        zorder=3
    )

    ctx.add_basemap(ax, source=ctx.providers.CartoDB.PositronOnlyLabels,
                    zoom='auto', zorder=4)

    plt.colorbar(im, ax=ax, label=cbar_label, shrink=0.6, pad=0.01)
    ax.set_title(title, fontsize=11, fontweight='bold')
    ax.set_axis_off()

plot_heatmap(
    ax_h, hxi, hyi, h_grid, hxmin, hxmax, hymin, hymax,
    title="Where Workers Live\n(avg commute distance from each home block)",
    cbar_label="Avg Commute Distance (miles)",
    md_boundary=md_state
)

plot_heatmap(
    ax_w, wxi, wyi, w_grid, wxmin, wxmax, wymin, wymax,
    title="Where Jobs Are Located\n(avg commute distance workers travel to reach each job center)",
    cbar_label="Avg Commute Distance (miles)",
    md_boundary=md_state
)

plt.tight_layout()
plt.savefig("commute_distance_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
print("Heatmap saved to commute_distance_heatmap.png")
```
